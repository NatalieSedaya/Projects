### Проверка рассчета эффективности обогащения на обучающей выборке

Эффективность обогощения рассчитывается по формуле:

Recovery = C (F-T) / F (C-T) * 100%, где:

- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

Используем в качестве показателей C, F, T столбцы rougher.output.concentrate_au, rougher.output.tail_au, rougher.input.feed_au: 

```python
def recovery(concentrate, feed, tail):
    final = concentrate*(feed - tail) / (feed*(concentrate - tail))*100
    
    return final

data_train['recovery_check'] = recovery(
    data_train['rougher.output.concentrate_au'], data_train['rougher.input.feed_au'], data_train['rougher.output.tail_au'])
```

```python
data_train['recovery_check'].describe()
```

Из-за отсутствия части данных, для рассчета МАЕ уберем строки с пропусками.

```python
print('МАЕ:' ,mean_absolute_error(data_train['rougher.output.recovery'] , data_train['recovery_check']))
```

Вывод:

Полученный МАЕ близок к 0, значит рассчет эффективности является правильным. Расхождения между расчетами и значениями признака rougher.output.recovery минимальны.

## Модель

### Функция для вычисления итоговой sMAPE

```python
def smape_above(target, predict):
    smape_a = 100/len(target) * np.sum(2 * np.abs(predict - target) / (np.abs(
        target) + np.abs(predict)))
    return smape_a
```

```python
def smape(target, predict):
    rougher = smape_above(target['rougher.output.recovery_x'], predict[:,0])
    final = smape_above(target['final.output.recovery_x'], predict[:,1])
    
    return (0.25 * rougher + 0.75 * final)
```
